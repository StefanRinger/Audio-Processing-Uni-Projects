   The value match operator `%in%` (see `?match`) can be helpful. Example: `a %in% b` returns `TRUE` for the elements of `a` being present within `b` and `FALSE` otherwise:

````r
#?subset
#?boxplot

#mushra

# if hidden reference rated lower than 90 points for >15% of occurrences -> remove the dude

hidref <- subset ( mushra , mushra $ responses_stimulus == "reference" )

id_list = unique(hidref $ questionaire_uuid)




for( i in 1:( length(id_list) ) ){

    
    reduced <- subset ( hidref $ responses_score , hidref $ questionaire_uuid == id_list[i] )
    
    temp <- (TRUE %in% (reduced < 90))
    
    #print(temp)
    
    if(temp==TRUE){
        id_list[i] <- ""
    }

    
}


clean_mushra <- subset ( mushra , mushra $ questionaire_uuid %in% id_list )

boxplot ( clean_mushra $ responses_score ~ clean_mushra $ responses_stimulus )


#id_list

#reduced <- subset ( mushra $ questionaire_uuid  mushra $ responses_score, mushra $ responses_stimulus == "reference" )

#hidref_disq <- subset ( mushra $ questionaire_uuid , mushra $ responses_stimulus == "reference" & mushra $ responses_score > 30 )

````

5. Generate a subset *mushracs* of the data that only includes the 3 different audio systems. Report if *responses_score* is normal distributed by providing both graphical analysis using Q-Q plots and statistical tests of normality. Check each codec seperately! Plot another boxplot with the final cleaned data.

````r
system_list <- c("Codec1","Codec2","Codec3")

#clean_mushra $ responses_stimulus == "Codec1"

clean_mushra <- subset ( clean_mushra , clean_mushra $ responses_stimulus %in% system_list )

#clean_mushra
clean_mushra1 <- subset ( clean_mushra , clean_mushra $ responses_stimulus %in% "Codec1" )

qqnorm(clean_mushra1 $ responses_score)
qqline(clean_mushra1 $ responses_score)

shapiro.test(clean_mushra1 $ responses_score)





clean_mushra2 <- subset ( clean_mushra , clean_mushra $ responses_stimulus %in% "Codec2" )

qqnorm(clean_mushra2 $ responses_score)
qqline(clean_mushra2 $ responses_score)

shapiro.test(clean_mushra2 $ responses_score)








clean_mushra3 <- subset ( clean_mushra , clean_mushra $ responses_stimulus %in% "Codec3" )

qqnorm(clean_mushra3 $ responses_score)
qqline(clean_mushra3 $ responses_score)

shapiro.test(clean_mushra3 $ responses_score)



# only the first Codec has some resemblance of a normal distributed value
# rest not normal distributed
````

Chi-Squared test can not be applied because the sample size is to small and samples don't follow a chi-squared distribution

````r
# calculate variability in each group

mushracs = clean_mushra
var_c1 = var(subset(mushracs, mushracs $ responses_stimulus == "Codec1") $ responses_score)
var_c2 = var(subset(mushracs, mushracs $ responses_stimulus == "Codec2") $ responses_score)
var_c3 = var(subset(mushracs, mushracs $ responses_stimulus == "Codec3") $ responses_score)

var_all = c(var_c1, var_c2, var_c3)
labels = c("Codec 1", "Codec 2", "Codec 3")
barplot(var_all ~ labels, xlab = "Codecs", ylab = "variance")

fligner.test(mushracs $ responses_score ~ mushracs $ responses_stimulus)
````
