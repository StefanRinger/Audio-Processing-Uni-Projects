```r
#Element wise logical comparison

cdc $ gender == "m"

cdc$age > 30
```

```r
# Now employ this as filter mask (subset)

# Extract only males

mdata <- subset ( cdc , cdc $ gender == "m" )
head ( mdata )



# AND conditions by &
m_and_over30 <- subset ( cdc , cdc $ gender == "m" & cdc $ age > 30 )
head ( m_and_over30 )

# OR conditions by |
m_or_over30 <- subset(cdc, cdc$gender == "m" | cdc$age > 30)
head ( m_or_over30 )
```

```r
# Box Plot of Height in both Men and Women
boxplot ( cdc $ height ~ cdc $ gender )

# The~ character can be read “versus” or “as a function of”. So we’re asking R to give us a box plots of heights where the groups are defined by gender.
```

```r
# calc the BMI
# all element wise...

bmi <- (cdc$weight/cdc$height^2) * 703
boxplot ( bmi ~ cdc $ genhlth )
# we see correlation of BMI and health...

# no correlation of BMI and smoking
boxplot ( bmi ~ cdc $ smoke100 )

# correlation of BMI and working out
boxplot ( bmi ~ cdc $ exerany )

```
